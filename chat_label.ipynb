# ðŸ§ª LLM-based Information Extraction from Real Estate Conversations

## ðŸ“‚ Dataset

The dataset consists of **20 conversational transcripts** between property buyers and real estate agents.  
These conversations are informal, multi-turn, and intentionally noisy, reflecting real-world interactions where information may be incomplete, implicit, or ambiguous.

Each conversation potentially contains details such as buyer identity, budget, location preferences, visit dates, profession, and buying timeline.  
A **manually curated gold label set** was created for each conversation to enable systematic evaluation.

---

## ðŸŽ¯ Problem Statement

Extracting structured information from conversational text is challenging due to missing fields, implicit references, and the risk of hallucination by LLMs.  
Unlike form-based inputs, conversational data requires careful handling to avoid incorrect inference while maintaining schema consistency.

---

## ðŸš€ Objective

The objective of this project is to evaluate how effectively **prompt engineering strategies** enable large language models to extract structured entities from conversational data.

Specifically, we aim to:
- Compare multiple prompting approaches
- Measure extraction quality using precision, recall, and F1 score
- Analyze model and entity-level failure modes

The focus is on **prompt design over model size**, reflecting practical constraints in real-world information extraction systems.

```python
groq_client = Groq(    api_key=os.getenv("GROQ_API_KEY"))

Models = ["llama-3.1-8b", "allam-7b", "llama-4-maverick-17b"]

GROQ_MODELS = {
    "llama-3.1-8b": "llama-3.1-8b-instant",
    "allam-7b": "allam-2-7b",
    "llama-4-maverick-17b": "meta-llama/llama-4-maverick-17b-128e-instruct"
}
```

```python
### Creating unified function to call llms and prompts and store in a single json file


import time

def call_llm(model_key: str, prompt: str):
    """
    Unified LLM caller for Phase 2 (Groq only).
    """
    start = time.time()

    response = groq_client.chat.completions.create(
        model=GROQ_MODELS[model_key],
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )

    latency = time.time() - start

    return (
        response.choices[0].message.content,
        latency,
        response.usage
    )
```

```python
results = []

for chat_id, conversation in chats.items():
    for model_key in Models:
        for prompt_name, prompt_template in PROMPTS.items():

            prompt = prompt_template.replace(
                "<<CONVERSATION>>", conversation
            )

            output, latency, tokens = call_llm(
                model_key=model_key,
                prompt=prompt
            )

            results.append({
                "chat_id": chat_id,
                "model": model_key,
                "prompt": prompt_name,
                "raw_output": output,
                "latency": latency,
                "token_usage": {"prompt_tokens": tokens.prompt_tokens,
                                "completion_tokens": tokens.completion_tokens,
                                "total_tokens": tokens.total_tokens} if tokens is not None else None
                                })
```

### Will check the output file, check its structure and then parse it obtain only our desired result.
### Then we will create the evaluation metric to check for the best performing LLM model and prompt strategy.
### As our main focus is to increase Precision (Reducing False Positive) it will be our most important metric to look after, then F1 score and then Recall.
### We will put our metric of all the models and prompt in a data frame, then aggregate the metrics to obtain the final result to finalise LLM and prompt.

```python
# Output is in string format, need to convert to dictionary

def safe_json_load(text):
    try:
        return json.loads(text)
    except Exception:
        return None

for r in res:
    r['parsed_output'] = safe_json_load(r['raw_output'])
```

```python
# Creating a function which will compare the parsed output with gold labels

def compare(pred, gold):
    if pred is None and gold is None:
        return "TN"
    if pred == gold:
        return "TP"
    if pred is not None and gold is None:
        return "FP"
    if pred is None and gold is not None:
        return "FN"
    return "FP"

# Creating a default dictionary for TP, TN, FP, FN counts

from collections import defaultdict

stats = defaultdict(lambda: {
    "TP": 0,
    "FP": 0,
    "FN": 0,
    "TN": 0
})
```

```python
# Compare model outputs with gold labels at the entity level.
# Invalid or malformed JSON outputs are skipped.
# For each valid response, predicted values are matched against
# ground truth and counted as TP / FP / FN / TN per model and prompt.
# These counts are later used to compute precision, recall, and F1.

for r in res:
   
    parsed = r["parsed_output"]
    if parsed is None:
        continue

    gold = gold_labels[r["chat_id"]]

    for field in gold:
        outcome = compare(parsed.get(field), gold.get(field))
        stats[(r["model"], r["prompt"], field)][outcome] += 1



# This block calculates the metrics(precison, recall and f1) for each model-prompt conmbination

def metrics(tp, fp, fn):
    precision = tp / (tp + fp) if (tp + fp) else 0
    recall = tp / (tp + fn) if (tp + fn) else 0
    f1 = (2 * precision * recall / (precision + recall)) if (precision + recall) else 0
    return precision, recall, f1
```

```python
# Creating metric table with information on TP, TN, FP, FN for the each model-prompt combination.



metrics_table = []

for (model, prompt, field), counts in stats.items():
    tp = counts["TP"]
    fp = counts["FP"]
    fn = counts["FN"]

    precision, recall, f1 = metrics(tp, fp, fn)

    metrics_table.append({
        "model": model,
        "prompt": prompt,
        "entity": field,
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "tp": tp,
        "fp": fp,
        "fn": fn
    })
```

```python
# Converting metrics table to dataframe for better visualization

import pandas as pd

df = pd.DataFrame(metrics_table)
df.head()
```

```python
# As discussed earlier, we will check first the collective performance of each model across all prompts.

model_summary = (df.groupby("model").agg({"f1": "mean","precision": "mean","recall": "mean"}).reset_index())

model_summary.sort_values("precision", ascending=False)
```

```python
# llama-4-maverick-17b model has failed to give json output for all the conversations hence it is not in the dataframe.

# Now checking the performance of each prompt across all models.

prompt_summary = (df.groupby("prompt").agg({"f1": "mean","precision": "mean","recall": "mean"}).reset_index())
prompt_summary.sort_values("precision", ascending=False)
```

```python
# Now checking for the model-prompt combination performance.

model_prompt_summary = (df.groupby(["model", "prompt"]).agg({"f1": "mean","precision": "mean","recall": "mean"}).reset_index())

model_prompt_summary.sort_values("precision", ascending=False)
```

```python
# Checking for the maximum hallucination made by each model-prompt combination for any entity.

max_hallucination = (df.groupby(["model", "prompt"]).agg({"fp": "max"}).reset_index()) 


max_hallucination.sort_values("fp", ascending=False)
```

Based on the experimental results, **Llama-3.1-8B** emerges as the most reliable model for conversational information extraction. It consistently achieved the **highest precision and F1 scores** while exhibiting **significantly fewer false positives**, indicating strong schema adherence and reduced hallucination behavior compared to Allam-7B.

Among prompt strategies, **null-bias and strict prompts** delivered the best overall performance. These prompts effectively constrained the model from inferring missing information, resulting in higher precision and more stable structured outputs.

In contrast, **Allam-7B** demonstrated higher hallucination rates across most prompt types, particularly when exposed to complex or constraint-heavy instructions. However, when paired with a **zero-shot prompt**, Allam-7B showed the **lowest hallucination levels**, making it a reasonable fallback option in **token-constrained or low-compute scenarios**, albeit with lower extraction accuracy.

Overall, the findings reinforce that **prompt discipline is critical for controlling hallucinations**, and that **stronger instruction-following models benefit more from constraint-based prompting**, while simpler prompts may be better suited for smaller or less robust models.
