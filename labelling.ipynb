```python
import os
import time
from dotenv import load_dotenv
from groq import Groq

load_dotenv()
```

```python
groq = Groq(api_key=os.getenv("GROQ_API_KEY"))

def chat_output(prompt: str) -> dict:
    """Function to get chat output from Groq API given a prompt string."""

    response = groq.chat.completions.create(model="llama-3.1-8b-instant", messages=[{"role": "user", "content": prompt}], temperature=0)

    return response.choices[0].message.content
```

```python
import os
from dotenv import load_dotenv
from groq import Groq
from openai import OpenAI

load_dotenv()

groq_client = Groq(
    api_key=os.getenv("GROQ_API_KEY")
)

openrouter_client = OpenAI(
    api_key=os.getenv("OPEN_ROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1"
)
```

```python
GROQ_MODELS = {
    "llama-3.1-8b": "llama-3.1-8b-instant",
    "llama-3.3-70b": "llama-3.3-70b-versatile"
}

OPENROUTER_MODELS = {
    "gpt-oss-120b": "openai/gpt-oss-120b:free",
    "gemma-3-27b": "google/gemma-3-27b-it:free"
}
```

```python
def call_llm(model_key: str, prompt: str):
    """
    Unified LLM caller for Phase 2.

    Args:
        model_key (str): logical model name
        prompt (str): full prompt text

    Returns:
        response_text (str)
        latency (float)
        token_usage (dict or None)
    """
    start = time.time()

    # ---- GROQ ----
    if model_key in GROQ_MODELS:
        response = groq_client.chat.completions.create(
            model=GROQ_MODELS[model_key],
            messages=[{"role": "user", "content": prompt}],
            temperature=0
        )

        output = response.choices[0].message.content
        tokens = response.usage

    # ---- OPENROUTER ----
    elif model_key in OPENROUTER_MODELS:
        response = openrouter_client.chat.completions.create(
            model=OPENROUTER_MODELS[model_key],
            messages=[{"role": "user", "content": prompt}],
            temperature=0
        )

        output = response.choices[0].message.content
        tokens = response.usage

    else:
        raise ValueError(f"Unsupported model_key: {model_key}")

    latency = time.time() - start

    return output, latency, tokens
```

```python
results = []

for chat_id, conversation in chats.items():
    for model in GROQ_MODELS:
        for prompt_name, prompt_template in PROMPTS.items():

            prompt = prompt_template.replace("<<CONVERSATION>>", conversation)

            response, latency, tokens = call_llm(
                model_key=model,
                prompt=prompt
            )

            results.append({
                "chat_id": chat_id,
                "model": model,
                "prompt": prompt_name,
                "raw_output": response,
                "latency": latency,
                "token_usage": tokens
            })
```
